# Clustering And Factorial analyses

First attempt at clustering (music characteristics and popularity)

```python
selected_columns = [
    'track_popularity', 'artist_popularity', 'danceability', 'energy', 'key',
    'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature'
]

df_selected = df[selected_columns]

print(df_selected.head())
```

```python
def detect_outliers_iqr(df):
    outlier_info = []
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
            percent = 100 * outliers / len(df)
            outlier_info.append({'feature': col, 'outliers': outliers, 'percent': percent})
    return pd.DataFrame(outlier_info).sort_values('percent', ascending=False)

outlier_stats = detect_outliers_iqr(df_selected)
print(outlier_stats)
```

RobustScaler for outlier heavy features, StandardScaler for the rest

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

outlier_heavy_features = ['instrumentalness', 'speechiness', 'acousticness']
standard_features = [
    'track_popularity', 'artist_popularity', 'danceability', 'energy', 'key',
    'loudness', 'mode', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature'
]

robust_pipe = Pipeline([('scaler', RobustScaler())])
standard_pipe = Pipeline([('scaler', StandardScaler())])

# Combine
scaler = ColumnTransformer([
    ('robust', robust_pipe, outlier_heavy_features),
    ('standard', standard_pipe, standard_features)
])

X_scaled = scaler.fit_transform(df_selected)

scaled_cols = outlier_heavy_features + standard_features
X_scaled_df = pd.DataFrame(X_scaled, columns=scaled_cols, index=df_selected.index)

print(X_scaled_df.head())

```

```python
from sklearn.decomposition import PCA

n_components_to_check = 5  
pca = PCA(n_components=n_components_to_check, random_state=42)
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)

print("Explained variance per PC:", explained_var)
print("Cumulative explained variance:", cumulative_var)
```

```python
plt.figure(figsize=(8,5))
plt.bar(range(1, n_components_to_check+1), explained_var, alpha=0.7, label='Individual')
plt.plot(range(1, n_components_to_check+1), cumulative_var, marker='o', color='red', label='Cumulative')
plt.xticks(range(1, n_components_to_check+1))
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

cluster_range = range(2, 11)
silhouette_scores = []
inertia_values = []

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, n_init=20, random_state=42)
    labels = kmeans.fit_predict(X_umap)
    
    inertia_values.append(kmeans.inertia_)
    score = silhouette_score(X_umap, labels)
    silhouette_scores.append(score)
    
    print(f"k={k}, silhouette score={score:.4f}")

# --- Plot silhouette scores ---
plt.figure(figsize=(8, 4))
plt.plot(cluster_range, silhouette_scores, 'o-', color='green')
plt.title('Silhouette Score vs. Number of Clusters (KMeans on UMAP)')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette score')
plt.grid(True)
plt.show()

# plot inertia (elbow method) ---
plt.figure(figsize=(8, 4))
plt.plot(cluster_range, inertia_values, 'o-', color='blue')
plt.title('Elbow Method (Inertia vs. k) on UMAP embedding')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
best_k = cluster_range[silhouette_scores.index(max(silhouette_scores))]

kmeans = KMeans(n_clusters=best_k, n_init=20, random_state=42)
final_labels = kmeans.fit_predict(X_umap)

sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=final_labels, palette='tab10', s=40)
plt.title(f'KMeans Clusters (k={best_k}) on UMAP Embedding')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```
